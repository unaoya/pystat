## 記述統計と推測統計

記述統計とは、集団の特徴を記述するため、すべての個体を観測し、データを整理要約し把握するものである。
例えば

- 国勢調査
- 会社の健康診断結果
- スポーツの記録
- 顧客の購買履歴

などを整理することで、対象となる集団全体の観測結果を把握する。

推測統計とは、集団の一部のみを観察し、全体の様子を推測するものである。
例えば

- 選挙速報（出口調査）
- 新製品のアンケート調査
- 視聴率

などがある。

全数調査が可能出ない場合とはどのような場合だろうか？
- コスト
- 破壊検査
- 未来過去のこと

## 母集団と標本
調査対象として想定する集団全体を母集団、そのうち分析のために選ばれた一部を標本という。

標本調査の考え方
標本にはばらつきがあるが、それをできるだけ取り除かなくては正しい推測を行うことができない。また、母集団の傾向をできるだけ再現するように標本を得ることが必要である。
そのための考え方として

- ランダム化
- 層化抽出

がある。

実際には統計分析にはバイアスがつきものである。
例えば世論調査の結果が新聞によって違うのはなぜだろうか？

標本調査の結果は偶然の産物だが、
標本を取り替えると何が起こるのかを考えるのが推測統計の理論が目指すところである。
この偶然性について取り扱うために、確率の考え方を援用する。

```python
import numpy as np

sample = np.random.normal(loc=170.0, scale=10.0, size=10)
np.mean(sample)
```

## データのばらつき、分散

上で見たように標本にはばらつきが存在する。
そのようなばらつきを扱うための数値的な指標が分散である。

詳しくは後ほど説明するが、分散を調べることで集団の傾向を知ることができたり、予測の精度について議論することができる。
統計や機械学習において、最も重要な量といってもよい。

## 機械学習と統計

機械学習と統計の違いとして、よく言われることは機械学習は予測がメイン、統計は説明がメインというものである。

実際、機械学習の手法ではまずは予測精度を優先にして、なぜそれがうまくいくのかの説明は後回しというものがある。

一方で、機械学習の手法やその正当性を説明するための道具として統計学が用いられる。

## モデルと予測

統計学や機械学習では、データを説明し、また予測するために、モデルを作る。
モデルは確率分布の言葉を用いて記述されることが多いが、表面上はそのように見えない場合もよくある。

モデルにはパラメータがあり、データを用いてパラメータを決定する。

色々なモデルを比較し、どのモデルがよりよく予測することができるかを調べていくことが必要になる。

## 数学との関係

この講義で扱う数学は主に線形代数、微積分、確率の三つに分けられる。
これら数学的な内容と統計、機械学習の関係について簡単に説明する。

### 線形代数
機械学習では多変数のデータを扱うが多変数のデータを扱うための基本的な数学的な方法が線型代数である。
統計や機械学習に現れる様々な手法やその計算は、ベクトルや行列といった線形代数の言葉を用いて表される。

線形とは？
幾何的には直線や平面といったまっすぐなもののこと。
関数の言葉では比例や一次関数のような単純な関係のこと。
機械学習では線形な手法が最も基本的である。
直線的な関係であるため人間の直感が働きやすい。
例えば線形回帰、線形判別といった手法がある。

### 微積分
微分はものの変化を調べるための方法。
特になんらかの量が最大や最小になるのはどのようなときかを調べる。

例えばデータを分析し予測をするときに、どのような式を用いて予測すると最もよく予測できるか？という問いを立てよう。
このとき、その式を決めるため、誤差が最小になるのはいつかを微分を用いて求める。

微分で行うことは、複雑なものの変化の比率を調べることにより、物事を線形化して調べやすくするということである。

積分は面積や密度の分布から重さを求めるためのもの。
積分でやっていることは、平たく言えば足し算である。

確率について数学的に扱うために用いられる。

### 確率
不確定要素を含む標本から母集団について推測する必要がある。
この不確定要素を数学的に表現するために用いられるのが確率である。
確率分布を用いてモデルを作るという作業を行う。
